```python
data1 = pd.read_csv("data_simple.csv", sep = '\t')
data2 = pd.read_csv("data_binary.csv", sep = '\t')
```

```python
n1 = data1.iloc[:, 0]
t1 = data1.iloc[:, 1]

n2 = data2.iloc[:, 0]
t2 = data2.iloc[:, 1]
```

```python
plt.scatter(n1, t1, 3, 'C1')

result1 = linregress(n1, t1)

x1 = np.linspace(min(n1), max(n1), 1000)
y1 = result1.slope * x1 + result1.intercept

plt.plot(x1, y1, color='C2')
```
