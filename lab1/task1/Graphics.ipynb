```python
data1 = pd.read_csv("data_simple_average.csv", sep = ' ')
data2 = pd.read_csv("data_binary_average.csv", sep = ' ')
data3 = pd.read_csv("data_simple_bad.csv", sep = ' ')
data4 = pd.read_csv("data_binary_bad.csv", sep = ' ')
```

```python
n1 = data1['size'].to_numpy()
t1 = data1['time'].to_numpy()

n2 = data2['size'].to_numpy()
t2 = data2['time'].to_numpy()

n3 = data3['size'].to_numpy()
t3 = data3['time'].to_numpy()

n4 = data4['size'].to_numpy()
t4 = data4['time'].to_numpy()
```

```python
def mapping_ln(x, a, b, c):
    return a * np.log(x - b) + c
```

```python
plt.scatter(n1, t1, 3, 'C1')

result1 = linregress(n1, t1)

x1 = np.linspace(min(n1), max(n1), 1000)
y1 = result1.slope * x1 + result1.intercept

plt.plot(x1, y1, color='C2')
plt.title("simple search average asimptotic")
plt.savefig("graphics/simple_average.png")
plt.show()

print(result1.slope)
```

```python
result2 = linregress(np.log10(n2), t2)

x2 = np.linspace(min(n2), max(n2), 1000)
y2 = result2.slope * np.log10(x2) + result2.intercept

plt.scatter(n2, t2, 3, 'C3')
plt.plot(x2, y2)

plt.title("binary search average asymptotic")
plt.savefig("graphics/binary_average.png")

plt.show()
```

```python
plt.scatter(n3, t3, 3, 'C1')

result3 = linregress(n3, t3)

x3 = np.linspace(min(n3), max(n3), 1000)
y3 = result3.slope * x3 + result3.intercept

plt.plot(x3, y3, color='C2')
plt.title("simple search bad asimptotic")
plt.savefig("graphics/simple_bad.png")
plt.show()
```

```python
plt.scatter((n4), (t4), 3, 'C3')

result4 = linregress((n4), (t4))

x4 = np.linspace(min((n4)), max((n4)), 1000)
y4 = result4.slope * x4 + result4.intercept

plt.plot(x4, y4)
plt.title("binary search bad asymptotic")
#plt.ylim(-1000, 2000)
plt.savefig("graphics/binary_bad.png")

plt.show()
```
