```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.optimize import curve_fit

import csv
```

```python
data_simple = pd.read_csv("data_simple.csv", sep = ' ')
data_linear = pd.read_csv("data_linear.csv", sep = ' ')
```

```python
size_simple = data_simple.iloc[:, 0]
time_simple = data_simple.iloc[:, 1]

plt.scatter(size_simple, (time_simple), 3)
slope, intercept, *_ = linregress(size_simple, (time_simple))

#x = np.linspace(min(size_simple), max(size_simple), 1000)
#y = slope * x + intercept
#plt.plot(x, y)

plt.show()
```

```python
size_linear = data_linear.iloc[:, 0]
time_linear = data_linear.iloc[:, 1]

plt.scatter(size_linear, time_linear, 3)

regress = linregress(size_linear, time_linear)

x = np.linspace(min(size_linear), max(size_linear), 1000)
y = regress.slope * x + regress.intercept

plt.plot(x, y)
plt.show()
```
